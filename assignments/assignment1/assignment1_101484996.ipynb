```python
"""
Author: KRISHNA VENU
Assignment: #1
"""

# b. Create 4 variables

# gym_member stores the name of the gym member as a string
gym_member = "Alex Alliton"  # str

# preferred_weight_kg stores the preferred weight in kilograms as a float
preferred_weight_kg = 20.5  # float

# highest_reps stores the highest number of repetitions achieved as an integer
highest_reps = 25  # int

# membership_active stores a boolean representing whether the membership is active
membership_active = True  # bool


# c. Create a dictionary named workout_stats
# workout_stats is a dictionary where the keys are the names of friends (strings),
# and the values are tuples containing the minutes spent on yoga, running, and weightlifting (integers).
workout_stats = {
    "Alex": (30, 45, 20),     # Yoga: 30 mins, Running: 45 mins, Weightlifting: 20 mins
    "Jamie": (25, 40, 30),    # Yoga: 25 mins, Running: 40 mins, Weightlifting: 30 mins
    "Taylor": (40, 50, 35)    # Yoga: 40 mins, Running: 50 mins, Weightlifting: 35 mins
}

# d. Using a loop, calculate the total workout minutes for each friend and add it to the dictionary
# Instead of modifying the dictionary during iteration, we store the total workout minutes in a separate dictionary
total_workout_minutes = {}

for friend, minutes in workout_stats.items():
    total_minutes = sum(minutes)  # Sum up the minutes of each activity for the total
    total_workout_minutes[f"{friend}_Total"] = total_minutes  # Store the total minutes separately

# Now, after the loop, we add the total minutes to the main dictionary
workout_stats.update(total_workout_minutes)

# e. Create a 2-dimensional (nested) list called workout_list
# workout_list is a list of lists, where each inner list contains the workout minutes for each friend. 
# The columns represent yoga, running, and weightlifting minutes.
workout_list = [
    list(workout_stats["Alex"][:3]),    # [Yoga: 30 mins, Running: 45 mins, Weightlifting: 20 mins]
    list(workout_stats["Jamie"][:3]),   # [Yoga: 25 mins, Running: 40 mins, Weightlifting: 30 mins]
    list(workout_stats["Taylor"][:3])   # [Yoga: 40 mins, Running: 50 mins, Weightlifting: 35 mins]
]

# f. Slice the workout_list to extract specific workout minutes for activities
# Extract and print the minutes for yoga and running for all friends
yoga_running_minutes = [row[:2] for row in workout_list]  # Extract columns 1 (Yoga) and 2 (Running)
print("Yoga and Running minutes for all friends:", yoga_running_minutes)

# Extract and print the minutes for weightlifting for the last two friends
weightlifting_minutes_last_two = [row[2] for row in workout_list[1:]]  # Extract column 3 (Weightlifting) for the last two friends
print("Weightlifting minutes for the last two friends:", weightlifting_minutes_last_two)

# g. Using an if-statement within a loop to check if any friend's total workout minutes >= 120
for friend in workout_stats:
    if "_Total" in friend:  # Look for friends with a '_Total' key (to get the total workout minutes)
        if workout_stats[friend] >= 120:  # If total workout minutes are >= 120
            print(f"Great job staying active, {friend.split('_')[0]}!")  # Split the '_Total' to get the friend's name and print the message


# h. Add a feature to allow the user to input a friend's name
friend_name = input("Enter a friend's name: ")

if friend_name in workout_stats:  # Check if the entered friend's name exists in the dictionary
    # If the name is found, print their workout minutes for each activity and their total workout minutes
    minutes = workout_stats[friend_name]
    total_minutes = sum(minutes)  # Calculate the total workout minutes for the friend

    # Print workout details for the specified friend
    print(f"\n{friend_name}'s workout minutes:")
    print(f"Yoga: {minutes[0]} minutes")
    print(f"Running: {minutes[1]} minutes")
    print(f"Weightlifting: {minutes[2]} minutes")
    print(f"Total workout minutes: {total_minutes} minutes")
else:
    # Print an error message if the friend's name is not in the records
    print(f"Friend {friend_name} not found in the records.")


# i. Find and print the friend with the highest and lowest total workout minutes
# First, we create a dictionary with friends' names and their total workout minutes
total_minutes_dict = {friend.split('_')[0]: workout_stats[friend] for friend in workout_stats if "_Total" in friend}

# Find the friend with the highest total workout minutes
friend_with_highest = max(total_minutes_dict, key=total_minutes_dict.get)

# Find the friend with the lowest total workout minutes
friend_with_lowest = min(total_minutes_dict, key=total_minutes_dict.get)

# Print the results
print(f"\nFriend with the highest total workout minutes: {friend_with_highest} ({total_minutes_dict[friend_with_highest]} minutes)")
print(f"Friend with the lowest total workout minutes: {friend_with_lowest} ({total_minutes_dict[friend_with_lowest]} minutes)")
```
